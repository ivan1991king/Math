Задание 1. Напишите код на Python, реализующий расчет длины вектора, заданного его координатами

```python
def len_x(x1,x2,x3):
    l_x = np.sqrt(x1**2+x2**2+x3**2)
    return l_x
```

Задание 3. Напишите код на Python, реализующий построение графиков:
окружности, эллипса, гиперболы.

```python
x = []
y1 = []
y2 = []
r=10
for i in range(-100,101):
    x_= r*i/100
    x.append(x_)
    y1.append(np.sqrt(r**2-x_**2))
    y2.append(-np.sqrt(r**2-x_**2))   

fig, ax = plt.subplots(1)
ax.plot(x,y1,'-b',x,y2,'-b')
plt.axis([-r, r, -r, r])   
ax.set_aspect(1)
plt.title('Окружность')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = []
y1 = []
y2 = []
a=4
b=10
for i in range(-100,101):
    x_= a*i/100
    x.append(x_)
    y1.append(np.sqrt(1-x_**2/a**2)*b)
    y2.append(-np.sqrt(1-x_**2/a**2)*b)
plt.axis([-a, a, -b, b])    
plt.plot(x,y1,'-b',x,y2,'-b')
plt.title('Эллипс')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = []
y1 = []
y2 = []
p=10
for i in range(1000):
    x_= i/1000
    x.append(x_)
    y1.append(np.sqrt(2*p*x_))
    y2.append(-np.sqrt(2*p*x_))   
plt.plot(x,y1,'-b',x,y2,'-b')
plt.title('Гипербола')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Задание 5. Нарисуйте трехмерный график двух параллельных плоскостей

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10,10,0.5)
Y = np.arange(-10,10,0.5)
X, Y = np.meshgrid(X,Y)
Z1 = 2*X + 4*Y
Z2 = 2*X + 4*Y + 200
ax.plot_wireframe(X,Y,Z1)
ax.plot_wireframe(X,Y,Z2)

show()
```
