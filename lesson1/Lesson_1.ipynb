Задание 4. Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), 
взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(0,10,121)
y1 = np.cos(x)
y2 = np.cos(4*x)
plt.axis([0, 10, -1, 1])
plt.plot(x, y1, x, y2)
plt.legend(('k=1','k=4'),loc='upper right')
plt.xlabel('x')
plt.ylabel('cos(kx)')
```
